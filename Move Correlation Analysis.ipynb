```python
# Import the packages

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read in the data
df = pd.read_csv(r'C:\Users\Kanne\Documents\movies.csv')
```

```python
#Read data in

df.head()
```

```python
# Find any missing data by looping through the columns

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(percent_missing*100)))
```

```python
# Data Types for columns

print(df.dtypes)
```

```python
#Change data types for columns

df["budget"] = df['budget'].astype('int')
df["gross"] = df['gross'].astype('int')
```

```python
#Add new column

df['Year'] = df['released'].astype(str).str[8:12]
df
```

```python
df.drop(columns=['Year'], axis = 1)
```

```python
# Find any Outliers?

df.boxplot(column=['gross'])
```

```python
#Remove any duplicates

df.drop_duplicates()
```

```python
# Order data 

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Plot budget and gross against each other

sns.regplot(x="gross", y="budget", data=df)
```

```python
#Gross vs Score

sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns
# Find all records containing 'pearson' and combine values

df.corr(method ='pearson')
```

```python
# Correlation Matrix between all numeric columns
# Find all records containing 'kendall' and combine values

df.corr(method ='kendall')
```

```python
# Correlation Matrix between all numeric columns
# Find all records containing 'spearman' and combine values

df.corr(method ='spearman')
```

```python
#Create correlation heatmap visualisation

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
